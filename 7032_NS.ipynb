# Stage 2 Generate Forecasts

```python
import sys
sys.path.append('D:/code/NS/Nonstationary_Transformers-main/')
# add the path to the time series library to import Autoformer model and time_features

import torch
import torch.fft
import os
from ns_models.ns_Transformer import Model
from utils.timefeatures import time_features
import argparse, json
import numpy as np
import shutil

class MY_MODEL():
    
    def __init__(self):
        """
        Load the trained model and all hyperparameters
        """
        self.model_path = './checkpoints/MSBA7032_short_84_21_ns_ns_Transformer_7032_1d_ftM_sl84_ll42_pl21_dm512_nh8_el2_dl1_df2048_fc1_ebtimeF_dtTrue_Exp_0/'
        self.para_path = './results/MSBA7032_short_84_21_ns_ns_Transformer_7032_1d_ftM_sl84_ll42_pl21_dm512_nh8_el2_dl1_df2048_fc1_ebtimeF_dtTrue_Exp_0/'
        self.pred_path = './results/'


        self.scaler_mean = np.load(self.para_path + 'scaler_mean.npy')  # load scaler_mean
        self.scaler_std = np.load(self.para_path + 'scaler_std.npy')    # load scaler_std
        self.stock_pred_list = np.load(self.pred_path + 'stock_pred_list.npy')   # load stock_pred_list

        print('loading parameters')
        config = json.load(open(self.para_path + 'config.json'))  # load configuration arguments 
        self.args = argparse.Namespace(**config)

        # Add default value for inverse attribute
        if not hasattr(self.args, 'inverse'):
            self.args.inverse = False

        self.timeenc = 0 if self.args.embed != 'timeF' else 1
        self.device = 'cpu'
        self.seq_len = self.args.seq_len
        self.label_len = self.args.label_len
        self.pred_len = self.args.pred_len
        self.return_type = '1d_next_styAdj'
        self.freq = self.args.freq

        print('loading model')
        self.my_model = Model(self.args).to(self.device)
        self.my_model.load_state_dict(torch.load(os.path.join(self.model_path, 'checkpoint.pth'), map_location=torch.device(self.device)))   # load trained model
        self.fcst = None

    def prepare_data(self, t):
        """
        Prepare data directly for time index t, following the Dataset_Custom class in data_loader.py
        """
        batch_data = self.dat.iloc[t-self.seq_len : t+self.pred_len]
        batch_data_oldstocks = pd.DataFrame(index=batch_data.index, columns=self.stock_pred_list, data=0)  # match stock_id with what we can predict from training data
        for s in self.stock_pred_list:
            batch_data_oldstocks[s] = batch_data.get(s, np.zeros(len(batch_data)))
        batch_data = batch_data_oldstocks.reset_index()

        # get mark data from time stamps, note we can use future time stamps (batch_y_mark) but cannot use future returns
        df_stamp = batch_data[['time']].rename(columns={'time':'date'})
        df_stamp['date'] = pd.to_datetime(df_stamp.date)
        if self.timeenc == 0:
            df_stamp['month'] = df_stamp.date.apply(lambda row: row.month, 1)
            df_stamp['day'] = df_stamp.date.apply(lambda row: row.day, 1)
            df_stamp['weekday'] = df_stamp.date.apply(lambda row: row.weekday(), 1)
            df_stamp['hour'] = df_stamp.date.apply(lambda row: row.hour, 1)
            data_stamp = df_stamp.drop(columns=['date']).values
        elif self.timeenc == 1:
            data_stamp = time_features(pd.to_datetime(df_stamp['date'].values), freq=self.freq)
            data_stamp = data_stamp.transpose(1, 0)

        # standardize time series based on trained mean and std
        batch_data = batch_data.iloc[:,1:].values
        if self.args.inverse:
            batch_data = (batch_data - self.scaler_mean) / self.scaler_std

        # input data is from the first seq_len points, output data is from seq_len-label_len to seq_len+pred_len
        batch_x = batch_data[0:self.seq_len]
        batch_y = np.concatenate((batch_data[self.seq_len-self.label_len : self.seq_len,:], np.zeros((self.pred_len,len(self.stock_pred_list)))), axis=0)  
        # batch_y does not use batch_data[seq_len-label_len:seq_len+pred_len,:] to make sure no look-ahead

        batch_x_mark = data_stamp[0:self.seq_len]
        batch_y_mark = data_stamp[self.seq_len-self.label_len : self.seq_len+self.pred_len]
        
        return batch_x, batch_x_mark, batch_y, batch_y_mark

    def gen_forecast_for(self, t):
        """
        Main function to generate forecast for time index t, following the test function in exp_long_term_forecasting.py
        """

        # get data for the current time
        batch_x, batch_x_mark, batch_y, batch_y_mark = self.prepare_data(t)  # prepare input data
        batch_x = torch.Tensor(batch_x).float().to(self.device).unsqueeze(0)
        batch_y = torch.Tensor(batch_y).float().to(self.device).unsqueeze(0)
        batch_x_mark = torch.Tensor(batch_x_mark).float().to(self.device).unsqueeze(0)
        batch_y_mark = torch.Tensor(batch_y_mark).float().to(self.device).unsqueeze(0)

        # decoder input
        dec_inp = torch.zeros_like(batch_y[:, -self.pred_len:, :]).float()
        dec_inp = torch.cat([batch_y[:, :self.label_len, :], dec_inp], dim=1).float().to(self.device)
        # encoder - decoder
        if self.args.use_amp:
            with torch.cuda.amp.autocast():
                outputs = self.my_model(batch_x, batch_x_mark, dec_inp, batch_y_mark)
        else:
            outputs = self.my_model(batch_x, batch_x_mark, dec_inp, batch_y_mark)

        f_dim = -1 if self.args.features == 'MS' else 0
        outputs = outputs[:, -self.pred_len:, :]
        #batch_y = batch_y[:, -self.pred_len:, :].to(self.device)
        outputs = outputs.detach().cpu().numpy()
        #batch_y = batch_y.detach().cpu().numpy()
        
        # inversly apply standardization
        if self.args.inverse:
            shape = outputs.shape
            outputs = (outputs.squeeze(0) * self.scaler_std + self.scaler_mean).reshape(shape)  # outputs dimension [1, pred_len, stock_pred_list]
            #batch_y = (batch_y.squeeze(0) * self.scaler_std + self.scaler_mean).reshape(shape)
        
        fcst_21d = np.sum(outputs, axis=1)[0,:]
        fcst_1d = outputs[0,0,:]

        return fcst_21d, fcst_1d
    
    def gen_forecasts(self, dat):
        """
        Generate forecasts for all time points in the input data if we can
        """
        self.master_index = dat[['stock_id']].copy()   # want to generate forecasts for these (time, stock_id) pairs
        self.time_points = self.master_index.index.unique()
        self.num_time_points = len(self.time_points)

        self.dat = dat.pivot_table(index='time', columns='stock_id', values=self.return_type).fillna(0.0) * 1000
        self.fcst = dat[['stock_id']].copy()
        self.fcst['forecast_21d'] = 0.0
        self.fcst['forecast_1d'] = 0.0
        
        self.my_model.eval()
        with torch.no_grad():
            for t in range(len(self.time_points)):
                tt = self.time_points[t]
                stocks = self.master_index.loc[tt,'stock_id'].values

                if (t < self.seq_len) or (t > self.num_time_points - self.pred_len):
                    pass
                else:
                    fcst_21d, fcst_1d = self.gen_forecast_for(t)
                    fcst = np.zeros(len(stocks))
                    fcst[np.isin(stocks, self.stock_pred_list)] = fcst_21d[np.isin(self.stock_pred_list, stocks)]
                    self.fcst.loc[tt, 'forecast_21d'] = fcst.copy().astype(float)
                    fcst = np.zeros(len(stocks))
                    fcst[np.isin(stocks, self.stock_pred_list)] = fcst_1d[np.isin(self.stock_pred_list, stocks)]
                    self.fcst.loc[tt, 'forecast_1d'] = fcst.copy().astype(float)

    def get_forecast_df(self):
        return self.fcst
```

# Stage III: Evaluation performance of forecasts

```python
def weighted_correlation(X,Y,weight_column):   

    # Calculate the weighted means
    mean_X = np.sum(X * weight_column) / np.sum(weight_column)
    mean_Y = np.sum(Y * weight_column) / np.sum(weight_column)

    # Calculate the weighted covariance
    weighted_covariance = np.sum(weight_column * (X - mean_X) * (Y - mean_Y))

    # Calculate the weighted standard deviations
    weighted_std_X = np.sqrt(np.sum(weight_column * (X - mean_X)**2))
    weighted_std_Y = np.sqrt(np.sum(weight_column * (Y - mean_Y)**2))

    # Calculate the weighted correlation coefficient
    weighted_correlation = weighted_covariance / (weighted_std_X * weighted_std_Y)

    return weighted_correlation
```

```python
import statsmodels.api as sm

def xs_neutral(dd, fcst_col):
    y = dd[fcst_col].values
    X = dd[['equal_weight','SIZE', 'SIZENL', 'LIQUIDITY', 'BETA', 'RESVOL',
            'MOMENTUM', 'LEVERAGE', 'VALUE', 'EARNYLD','GROWTH', 'LTREV', 'STREV']].values   ## market factor = equal_weight
    w = dd['liq_weight'].values

    mod_wls = sm.WLS(y, X, weights=w).fit()
    res = y - mod_wls.fittedvalues
    return pd.Series(data=res, index=dd.index)

df['forecast_21d_neutral'] = df.groupby(df.index).apply(lambda x: xs_neutral(x,'forecast_21d')).values
df['forecast_1d_neutral'] = df.groupby(df.index).apply(lambda x: xs_neutral(x,'forecast_1d')).values
```

```python
res = pd.DataFrame(index=['forecast_1d_neutral','forecast_21d_neutral'], columns=['IC_21d','IC_10d','IC_5d','IC_1d','Annual Mean (%)','Annual Vol (%)','Annual SR','Turnover'], data=0)
res = res.astype(float)

weight = df['liq_weight'].values   # always use liq_weight in evaluation
for hz in [21, 10, 5, 1]:   # ['21d_next_mktAdj', '10d_next_mktAdj', '5d_next_mktAdj', '1d_next_mktAdj']
    mr = '%sd_next_mktAdj' % hz
    target = df[mr].values
    res.loc['forecast_1d_neutral','IC_%sd' % hz] = weighted_correlation(df['forecast_1d_neutral'].values, target, weight)
    res.loc['forecast_21d_neutral','IC_%sd' % hz] = weighted_correlation(df['forecast_21d_neutral'].values, target, weight)

for y_col in ['forecast_1d_neutral','forecast_21d_neutral']:
    y_hat = df[y_col].values
    portfolio = df[['stock_id','1d_next_mktAdj','liq_weight']].copy()
    portfolio['pos'] = weight * y_hat
    portfolio['pos'] = portfolio.groupby(portfolio.index)['pos'].transform(lambda x: x/np.sum(np.abs(x)))
    pnl = portfolio.groupby(portfolio.index).apply(lambda x: np.sum(x['pos'] * x['1d_next_mktAdj']))
    portfolio['prev_pos'] = portfolio.groupby(portfolio.stock_id)['pos'].shift().fillna(0.0)
    turnover = portfolio.groupby(portfolio.index).apply(lambda x: np.sum(np.abs(x['pos']-x['prev_pos'])))

    res.loc[y_col,'Annual Mean (%)'] = pnl.mean() * 244 * 100
    res.loc[y_col,'Annual Vol (%)'] = pnl.std() * np.sqrt(244) * 100
    res.loc[y_col,'Annual SR'] = pnl.mean() / pnl.std() * np.sqrt(244)
    res.loc[y_col,'Turnover'] = turnover.mean()

if not os.path.exists('performance'):
    os.makedirs('performance')

# res.to_csv('performance/vali_results8421.csv')
display(res)
```

```python
# single time cross-sectional neutralization

import statsmodels.api as sm

def xs_neutral(dd, fcst_col):
    y = dd[fcst_col].values
    X = dd[['equal_weight','SIZE', 'SIZENL', 'LIQUIDITY', 'BETA', 'RESVOL',
            'MOMENTUM', 'LEVERAGE', 'VALUE', 'EARNYLD','GROWTH', 'LTREV', 'STREV']].values
    w = dd['liq_weight'].values

    mod_wls = sm.WLS(y, X, weights=w).fit()
    res = y - mod_wls.fittedvalues
    return pd.Series(data=res, index=dd.index)

df['forecast_21d_neutral'] = df.groupby(df.index).apply(lambda x: xs_neutral(x,'forecast_21d')).values
df['forecast_1d_neutral'] = df.groupby(df.index).apply(lambda x: xs_neutral(x,'forecast_1d')).values
```

```python
res = pd.DataFrame(index=['forecast_1d_neutral','forecast_21d_neutral'], columns=['IC_21d','IC_10d','IC_5d','IC_1d','Annual Mean (%)','Annual Vol (%)','Annual SR','Turnover'], data=0)
res = res.astype(float)

weight = df['liq_weight'].values   # always use liq_weight in evaluation
for hz in [21, 10, 5, 1]:   # ['21d_next_mktAdj', '10d_next_mktAdj', '5d_next_mktAdj', '1d_next_mktAdj']
    mr = '%sd_next_mktAdj' % hz
    target = df[mr].values
    res.loc['forecast_1d_neutral','IC_%sd' % hz] = weighted_correlation(df['forecast_1d_neutral'].values, target, weight)
    res.loc['forecast_21d_neutral','IC_%sd' % hz] = weighted_correlation(df['forecast_21d_neutral'].values, target, weight)

for y_col in ['forecast_1d_neutral','forecast_21d_neutral']:
    y_hat = df[y_col].values
    portfolio = df[['stock_id','1d_next_mktAdj','liq_weight']].copy()
    portfolio['pos'] = weight * y_hat
    portfolio['pos'] = portfolio.groupby(portfolio.index)['pos'].transform(lambda x: x/np.sum(np.abs(x)))
    pnl = portfolio.groupby(portfolio.index).apply(lambda x: np.sum(x['pos'] * x['1d_next_mktAdj']))
    portfolio['prev_pos'] = portfolio.groupby(portfolio.stock_id)['pos'].shift().fillna(0.0)
    turnover = portfolio.groupby(portfolio.index).apply(lambda x: np.sum(np.abs(x['pos']-x['prev_pos'])))

    res.loc[y_col,'Annual Mean (%)'] = pnl.mean() * 244 * 100
    res.loc[y_col,'Annual Vol (%)'] = pnl.std() * np.sqrt(244) * 100
    res.loc[y_col,'Annual SR'] = pnl.mean() / pnl.std() * np.sqrt(244)
    res.loc[y_col,'Turnover'] = turnover.mean()

# res.to_csv('performance/test_results8405.csv')
display(res)
```
